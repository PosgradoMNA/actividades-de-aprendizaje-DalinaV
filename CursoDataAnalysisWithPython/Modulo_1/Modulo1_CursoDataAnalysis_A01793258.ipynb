El primer atributo, "simbolización", corresponde al nivel de riesgo del seguro de un coche.
A los coches se les asigna inicialmente un símbolo de factor de riesgo asociado a su precio.
Luego, si un automóvil es más arriesgado, este símbolo se ajusta desplazándolo hacia arriba en la escala.
Un valor de +3 indica que el coche es arriesgado, -3 que probablemente es bastante seguro.
El segundo atributo, "pérdidas normalizadas", es el promedio relativo de pagos por siniestro por año del vehículo asegurado.
año del vehículo asegurado.
Este valor se normaliza para todos los coches de una determinada clasificación de tamaño (dos puertas

coche al año.
Los valores van de 65 a 256.
Los demás atributos son fáciles de entender.
Bien, después de entender el significado de cada característica, nos daremos cuenta de que el 26º atributo
es "precio".
Este es nuestro valor objetivo, o etiqueta, en otras palabras.
Esto significa que "precio" es el valor que queremos predecir del conjunto de datos, y los
predictores deberían ser todas las demás variables enumeradas, como "simbolización", "pérdidas normalizadas"
